```python
import networkx as nx
import matplotlib.pyplot as plt

# Crear un grafo vacío
G = nx.Graph()

# Definir los routers y sus interfaces
routers = {
    "R1": ["10.0.0.1", "192.168.0.1"],
    "R2": ["10.0.0.2", "192.168.1.2", "192.168.4.1", "192.168.5.1", "192.168.10.2"],
    "R3": ["10.0.0.3", "192.168.2.2", "192.168.5.2", "192.168.9.3"],
    "R4": ["10.0.0.4", "192.168.3.2", "192.168.8.2"],
    "R5": ["10.0.0.5", "192.168.4.2", "192.168.6.2"]
}

# Agregar nodos (routers) al grafo
for router in routers:
    G.add_node(router)

# Definir los enlaces y sus atributos
links = [
    ("R1", "R2", "eth0", "eth0", {"metric": 2}),
    ("R2", "R3", "eth1", "eth0", {"metric": 10}),
    ("R2", "R5", "eth2", "eth1", {"metric": 1}),
    ("R3", "R4", "eth1", "eth1", {"metric": 8}),
    ("R3", "R5", "eth2", "eth0", {"metric": 4}),
    ("R4", "R2", "eth0", "eth3", {"metric": 6})
]

# Agregar enlaces al grafo
for link in links:
    src_router, dest_router, src_intf, dest_intf, attrs = link
    G.add_edge(src_router, dest_router, **attrs)

# Dibujar el grafo
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, font_weight='bold')
nx.draw_networkx_edge_labels(G, pos)
plt.show()
```
